```python
#prewitt and sobel on one image

import cv2
import numpy as np


def prewittAndSobel (image, horizontal, vertical):
    
    # Apply gray scale
    gray_img = np.round(0.2989 * img[:, :, 0] +
                        0.5870 * img[:, :, 1] +
                        0.1140 * img[:, :, 2]).astype(np.uint8)

    height, width = gray_img.shape

    newgradientImage = np.zeros((h, w))

    for i in range(1, height - 1):
        for j in range(1, width - 1):
            horizontalGrad = (horizontal[0, 0] * gray_img[i - 1, j - 1]) + \
                             (horizontal[0, 1] * gray_img[i - 1, j]) + \
                             (horizontal[0, 2] * gray_img[i - 1, j + 1]) + \
                             (horizontal[1, 0] * gray_img[i, j - 1]) + \
                             (horizontal[1, 1] * gray_img[i, j]) + \
                             (horizontal[1, 2] * gray_img[i, j + 1]) + \
                             (horizontal[2, 0] * gray_img[i + 1, j - 1]) + \
                             (horizontal[2, 1] * gray_img[i + 1, j]) + \
                             (horizontal[2, 2] * gray_img[i + 1, j + 1])

            verticalGrad = (vertical[0, 0] * gray_img[i - 1, j - 1]) + \
                           (vertical[0, 1] * gray_img[i - 1, j]) + \
                           (vertical[0, 2] * gray_img[i - 1, j + 1]) + \
                           (vertical[1, 0] * gray_img[i, j - 1]) + \
                           (vertical[1, 1] * gray_img[i, j]) + \
                           (vertical[1, 2] * gray_img[i, j + 1]) + \
                           (vertical[2, 0] * gray_img[i + 1, j - 1]) + \
                           (vertical[2, 1] * gray_img[i + 1, j]) + \
                           (vertical[2, 2] * gray_img[i + 1, j + 1])

            # Edge Magnitude
            mag = np.sqrt(pow(horizontalGrad, 2.0) + pow(verticalGrad, 2.0))
            newgradientImage[i - 1, j - 1] = mag
    
    return newgradientImage

image = cv2.imread("/Users/jasleenkaur/Desktop/Masters/Spring 2022/Image Analysis/Project/Cancerous cell smears/cyl01.bmp")

prewittHorizontal = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
prewittVertical = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])

sobelHorizontal = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]]) 
sobelVertical = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])

newImage = prewittAndSobel(image, prewittHorizontal, prewittVertical)
cv2.imwrite("/Users/jasleenkaur/Desktop/Masters/Spring 2022/Image Analysis/Project/Sobel/Prewittcyl01.bmp", newImage)

newImage = prewittAndSobel(image, sobelHorizontal, sobelVertical)
cv2.imwrite("/Users/jasleenkaur/Desktop/Masters/Spring 2022/Image Analysis/Project/Sobel/Sobelcyl01.bmp", newImage)

```
